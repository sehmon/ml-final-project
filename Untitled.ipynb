We'll be using the Million Song Dataset to try and predict a track's popularity given it's metadata. Let's first start with our imports.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from graphviz import *
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import Imputer
from sklearn.tree import export_graphviz
from sklearn.model_selection import KFold
from sklearn import tree

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

Let's first import our data file and plot the features to see what would be the best features to split on.

```python
music_data = pd.read_csv('music.csv')
music_data.hist()
plt.show()
```

Now let's remove the columns we find extraneous to our dataset. Removing these columns will reduce the noise our classifier faces.

```python
cols = [
    'artist.hotttnesss',
    #'artist.id',
    #'artist.name',
    #'artist_mbtags',
    #'artist_mbtags_count',
    #'bars_confidence',
    'bars_start',
    #'beats_confidence',
    'beats_start',
    'duration',
    #'end_of_fade_in',
    'familiarity',
    'key',
    #'key_confidence',
    #'latitude',
    #'location',
    #'longitude',
    'loudness',
    'mode',
    #'mode_confidence',
    #'release.id',
    #'release.name',
    #'similar',
    'song.hotttnesss',
    #'song.id',
    #'start_of_fade_out',
    #'tatums_confidence',
    'tatums_start',
    'tempo',
    #'terms',
    #'terms_freq',
    'time_signature',
    #'time_signature_confidence',
    #'title',
    'year'
]
music_data_dropped = music_data[cols]
music_data_dropped = music_data_dropped.fillna(0)
```

Now we can plot the distribution of data we have left.

```python
music_data_dropped.hist()
plt.show()
```

```python
X = music_data_dropped.drop('song.hotttnesss', axis=1)  
y = music_data_dropped['song.hotttnesss']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)

classifier = DecisionTreeRegressor(max_depth=7)
classifier.fit(X=X_train, y=y_train)
```

```python
y_pred = classifier.predict(X_test)
classifier.score(X_test, y_test)
```

```python
export_graphviz(classifier, out_file='tree.dot', feature_names=list(X_test))
```

Now that we have a working classifier, let's try to experiment and improve our score

## Model Experiments

Let's first try writing some code to do 10-Fold cross validation

```python
kf = KFold(n_splits=10)
kf.get_n_splits(music_data_dropped)
```

```python
score_list = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    classifier = DecisionTreeRegressor(max_depth=5)
    classifier.fit(X=X_train, y=y_train)
    y_pred = classifier.predict(X_test)
    score_list.append(classifier.score(X_test, y_test))

sum(score_list) / float(len(score_list))
    
```

```python
export_graphviz(classifier, out_file='tree.dot', feature_names=list(X_test))
```
